## 코인데스크 기사 크롤링

### 라이브러리 Import

```python
import requests
from bs4 import BeautifulSoup
```

### 크롤링 URL 주소

```python
base_url = 'http://www.coindeskkorea.com'
uri = '/news/articleList.html?view_type=sm'
url = base_url + uri
```

```python
url
```

### 세션 연결

```python
session = requests.session()
session
```

### 응답받기

```python
response = session.get(url)
response
```

### HTML을 parsing

```python
html = response.text
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup
```

### result에 Data 담기

```python
item_number = len(soup.select('div.text-block > div.list-titles > a > strong'))
```

```python
result = []
for i in range(0, item_number):
    subject = soup.select('div.text-block > div.list-titles > a > strong')[i].get_text().strip()
    summary = soup.select('div.text-block > p > a')[i].text.strip()
    url = base_url + soup.select('div.text-block > p > a')[i].attrs['href']
    author = soup.select('div.text-block > div.list-dated > strong')[i].text.strip()

    article = {
        'subject': subject,
        'summary': summary,
        'url': url,
        'author': author
    }
    result.append(article)
```

```python
url = 'http://www.coindeskkorea.com/news/articleList.html?page=2&total=4864&box_idxno=&view_type=sm'
```

```python
result = []
for page_num in range(1, 11):
    url = 'http://www.coindeskkorea.com/news/articleList.html?page='+ str(page_num) +'&total=4864&box_idxno=&view_type=sm'
    response = session.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    for i in range(0, item_num):
        subject = soup.select('div.text-block > div.list-titles > a > strong')[i].get_text().strip()
        summary = soup.select('div.text-block > p > a')[i].text.strip()
        url = base_url + soup.select('div.text-block > p > a')[i].attrs['href']
        author = soup.select('div.text-block > div.list-dated > strong')[i].text.strip()

        article = {
            'subject' : subject,
            'summary': summary,
            'url': url,
            'author': author
        }

        result.append(article)
        
```
