### 감성사전구축

- 데이터 : https://archive.ics.uci.edu/ml/datasets/Sentiment+Labelled+Sentences

```python
import pandas as pd
```

```python
df=pd.read_csv("amazon_cells_labelled.txt",sep="\t",header=None)
df
```

```python
df.info()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
X=df[0]
y=df[1]
```

```python
tfidf=TfidfVectorizer(stop_words='english',max_features=1000)
X_tdm=tfidf.fit_transform(X)
```

```python
X_tdm.toarray()
```

```python
tfidf.get_feature_names()
```

```python
tfidf.inverse_transform(X_tdm[0])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
x_train,x_test,y_train,y_test=train_test_split(X_tdm,y,test_size=0.3,random_state=103)

print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
lr_cif=LogisticRegression()
lr_cif.fit(x_train,y_train)
```

```python
lr_cif.score(x_train,y_train)
```

```python
pred=lr_cif.predict(x_test)
```

```python
accuracy_score(y_test,pred)
```

```python
lr_cif.coef_
```

### < 과제 >
- TFIDF 이용해서 네이버 영화 평점 감성 사전 구축
    - ratings.txt 
- 극성 구하기


```python
df=pd.read_csv("ratings.txt",sep="\t",header=None)
df
```

```python
df.info()
```

```python
X=df[1]
y=df[2]
```

```python
X =X.str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
```

```python
X
```

```python
X = X.str.replace('^ +', "") 
X.replace('', np.nan, inplace=True)
print(X.isnull().sum())
```

```python
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']
```

```python
X
```

```python
from konlpy.tag import Okt

okt = Okt()


def tw_tokenizer(text):
    tokens_ko = okt.morphs(text)
    return tokens_ko
```

```python
tfidf=TfidfVectorizer(tokenizer=tw_tokenizer,stop_words=stopwords,min_df=1,sublinear_tf=True )
X = tfidf.fit_transform(df[1].values.astype('U'))
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=103)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
lr_cif=LogisticRegression()
lr_cif.fit(x_train,y_train)
lr_cif.score(x_train,y_train)
pred=lr_cif.predict(x_test)
accuracy_score(y_test,pred)
```

```python
lr_cif.coef_
```

```python
st_df=pd.DataFrame({"단어":tfidf.get_feature_names(),"회귀계수":lr_cif.coef_.flat})
st_df.tail()
```

```python
st_neg=st_df[st_df["회귀계수"]<0].sort_values("회귀계수")
st_neg.head()
```

```python
st_pos=st_df[st_df["회귀계수"]>0].sort_values("회귀계수",ascending=False)
st_pos.head()
```

```python
import numpy as np

st_df["극성"]=np.sign(st_df["회귀계수"])

st_df.tail()
```

```python
st_df["극성"].value_counts()
```
